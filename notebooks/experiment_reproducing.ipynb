# MSD Dataset

```python
path = "outputs/MDS_benchmark_all_methods.csv"
data = pd.read_csv(path)
data.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

plt.figure(figsize=(7, 5))

sns.lineplot(
    data=data,
    x="time_sec",
    y="rel_err",
    hue="method",
    style="method",
    markers=True,
    dashes=False,
    linewidth=2.5,
    markersize=7,
    alpha=0.95
)

plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Relative error", fontsize=12)

plt.title("Accuracy vs Runtime for Kernel Solvers", fontsize=13, weight="bold")

plt.legend(
    title="Method",
    fontsize=10,
    title_fontsize=11,
    frameon=True,
    fancybox=True,
    framealpha=0.95
)
plt.tight_layout()
```
